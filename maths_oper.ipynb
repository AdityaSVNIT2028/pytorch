```python
import torch
import numpy as np
```

```python
# in this we look around the math operations in this
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
tensor_a + tensor_b
```

```python
torch.add(tensor_a, tensor_b)
```

```python
#substraact
tensor_a - tensor_b
```

```python
torch.sub(tensor_a, tensor_b)
```

```python
# multiplication
torch.mul(tensor_a, tensor_b)
```

```python
#division
tensor_b/ tensor_a
```

```python
torch.div(tensor_b, tensor_b)
```

```python
# now the remainder
torch.remainder(tensor_b, tensor_a)
```

```python
tensor_b % tensor_a
```

```python
# for exponent and power
tensor_a**tensor_b
```

```python
torch.pow(tensor_a, tensor_b)
```

```python
print(tensor_a)
tensor_a.add (tensor_b)
print(tensor_a)
```

```python
# here one important thing to note that if we are usign
# tensor_a.add() and tensor_a.add_() then by 2nd we directly
# reassign
tensor_a.add_(tensor_b)
print(tensor_a )
```
