```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
#reshape and View
m_torch = my_torch.reshape(5,2)
m_torch
```

```python
# Reshape if we don't know the number of items using -1
my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
# the above one is valid for either one between the row and column is known and the unknown one is replaced by -1
# for ex:
my_torch2 = my_torch2.reshape(-1,2)
my_torch2
```

```python
# now for using the view
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(5,2)
my_torch4
```

```python
# with reshape and view they will update
my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 21
my_torch5
```

```python
my_torch6
```
